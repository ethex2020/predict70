```python
import os
from keras import backend as K
from keras.preprocessing.image import ImageDataGenerator


img_rows, img_cols = 150, 150
img_channels = 3

validation_steps = 2000   # 0 till 2000

batch_size_Tatal_sample_size = 20 

base_dir = '/media/ssd/Windows7_1/1_ssd_m.2/PetImages' 
## base_dir = 'I:\\1_ssd_m.2\\PetImages' 

validation_dir = os.path.join(base_dir, 'validation')
print("Finished dir read")

validation_datagen = ImageDataGenerator(rescale=1./255)

validation_generator = validation_datagen.flow_from_directory(
        validation_dir,
        target_size=(img_rows, img_cols),
        batch_size=batch_size_Tatal_sample_size,
        class_mode='categorical')


#valida_x, valida_y = validation_generator.next()
#validation_x, validation_y = validation_generator.next()

result = model.evaluate_generator(validation_generator, steps=validation_steps, verbose=1)

for name, value in zip(model.metrics_names, result):
    print(name, value)
   
print("{0}: {1:.2%}".format(model.metrics_names[1], result[1]))
```
